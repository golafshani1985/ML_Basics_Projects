```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic_processed.csv')
df.head()
```

```python
df.describe(include='all')
```

```python
survival_rate = df['Survived'].mean() * 100
print(f"Survival Rate: {survival_rate:.2f}%")
```

```python
gender_survival = df.groupby('Sex')['Survived'].mean() * 100
print(gender_survival)
```

```python
sns.barplot(x=gender_survival.index, y=gender_survival.values)
plt.ylabel('Survival Rate (%)')
plt.title('Survival Rate by Gender')
plt.show()
```

```python
Pclass_survival = df.groupby('Pclass')['Survived'].mean() * 100
print('Pclass__survival')
```

```python
sns.barplot(x=Pclass_survival.index, y=Pclass_survival.values)
plt.title('Survival Rate by Pclass(%)')
plt.xlabel('Class')
plt.ylabel('Survival Rate (%)')
plt.show()
```

```python

sns.histplot( data= df,x='Age',bins=20)
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```
